# Standardization

```python
df=df.drop(columns=['sex','smoker','day','time'])
```

```python
df.head()
```

```python
x=df.drop(columns={'size'},axis=1) #independent column
y=df['size']
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
print(df.shape)
print(x.shape)
print(x_train.shape)
print(x_test.shape)
print(y.shape)
print(y_train.shape)
print(y_test.shape)
```
